```python
import nltk

# Define the sentence to be parsed
sentence = " the cat chased the mouse "

# Define the PCFG grammar
grammar = nltk.PCFG.fromstring("""
    S -> NP VP [1.0]
    NP -> Det N [0.5] 
    NP -> NP PP [0.2] 
    NP -> 'I' [0.1]
    NP -> 'you' [0.1]
    NP -> 'he' [0.1]
    VP -> V NP [0.5]
    VP ->VP PP [0.2]
    VP ->'said' [0.1]
    VP ->'ate' [0.1]
    VP ->'chased' [0.1]
    PP -> P NP [1.0]
    Det -> 'the' [0.6]
    Det -> 'a' [0.4]
    N -> 'cat' [0.4]
    N -> 'mouse' [0.3]
    N -> 'dog' [0.3]
    V -> 'chased' [0.4]
    V -> 'ate' [0.3] 
    V -> 'saw' [0.3]
    P -> 'with' [0.4]
    P -> 'in' [0.3] 
    P -> 'on' [0.3]
""")

#Viterbi parser uses CYK algo

def cyk_algo(sentence,grammar):
    tokens=sentence.split()
    parser=nltk.ViterbiParser(grammar)
    
    return list(parser.parse(tokens))

for tree in cyk_algo(sentence,grammar):
    print(tree)
```
